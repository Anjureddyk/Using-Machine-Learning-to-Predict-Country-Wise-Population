```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("population_by_country_2020.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
data.describe
```

```python
data.describe()
```

```python
data['Migrants (net)'].fillna(0, inplace=True)
```

```python
data.isnull().sum()
```

```python
# Remove the percentage sign and convert 'Yearly Change' column to float type
data['Yearly Change'] = data['Yearly Change'].str.rstrip('%').astype(float)
```

```python
data.describe()
```

```python
data.isnull
```

```python
# Remove the percentage sign and convert 'World Share' column to float type
data['World Share'] = data['World Share'].str.rstrip('%').astype(float)
```

```python
# Convert 'Med. Age' column to float type
# Replace 'N.A.' values with NaN
data['Med. Age'].replace('N.A.', np.nan, inplace=True)

# Convert 'Med. Age' column to float type
data['Med. Age'] = data['Med. Age'].astype(float)
```

```python
# Convert 'Fert. Rate' column to float type
# Replace 'N.A.' values with NaN
data['Fert. Rate'].replace('N.A.', np.nan, inplace=True)

# Convert 'Fert. Rate' column to float type
data['Fert. Rate'] = data['Fert. Rate'].astype(float)
```

```python
# Convert 'Urban Pop %' column to float type
# Replace 'N.A.' values with NaN
data['Urban Pop %'].replace('N.A.', np.nan, inplace=True)

# Remove the percentage sign and convert 'World Share' column to float type
data['Urban Pop %'] = data['Urban Pop %'].str.rstrip('%').astype(float)
```

```python
data.isnull().sum()
```

```python
# Fill NaN values in 'Fert. Rate' column with the mean
data['Fert. Rate'].fillna(data['Fert. Rate'].mean(), inplace=True)
```

```python
data.isnull().sum()
```

```python
# Fill NaN values in 'Med. Age' column with the mean
data['Med. Age'].fillna(data['Med. Age'].mean(), inplace=True)
```

```python
data.isnull().sum()
```

```python
# Fill NaN values in 'Urban Pop %' column with the mean
data['Urban Pop %'].fillna(data['Urban Pop %'].mean(), inplace=True)
```

```python
data.isnull().sum()
```

```python
data.to_csv("pre_processed_data.csv")
```

```python
data = pd.read_csv("pre_processed_data.csv")
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar Plot: Population of Each Country
plt.figure(figsize=(10, 6))
sns.barplot(x='Country (or dependency)', y='Population', data=data)
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Population of Each Country')
plt.xticks(rotation=45)
plt.show()
```

```python
# Line Plot: Yearly Change in Population
plt.figure(figsize=(10, 6))
sns.lineplot(x='Country (or dependency)', y='Yearly Change', data=data)
plt.xlabel('Country')
plt.ylabel('Yearly Change')
plt.title('Yearly Change in Population')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter Plot: Net Change in Population vs. Density
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Net Change', y='Density (P/Km²)', data=data)
plt.xlabel('Net Change')
plt.ylabel('Density (P/Km²)')
plt.title('Net Change in Population vs. Density')
plt.show()
```

```python
# Histogram: Distribution of Population Density
plt.figure(figsize=(10, 6))
sns.histplot(data['Density (P/Km²)'], bins=10, kde=True)
plt.xlabel('Density (P/Km²)')
plt.ylabel('Count')
plt.title('Distribution of Population Density')
plt.show()
```

```python
# Box Plot: Distribution of Fertility Rates
plt.figure(figsize=(10, 6))
sns.boxplot(data['Fert. Rate'])
plt.xlabel('Fertility Rate')
plt.title('Distribution of Fertility Rates')
plt.show()
```

```python
# Pie Chart: Urban Population Percentage for Each Country
plt.figure(figsize=(10, 6))
plt.pie(data['Urban Pop %'], labels=data['Country (or dependency)'], autopct='%1.1f%%')
plt.title('Urban Population Percentage for Each Country')
plt.show()
```

```python
# Bar Plot: Top 5 Countries with the Highest Net Migration
plt.figure(figsize=(10, 6))
top_5_countries = data.nlargest(5, 'Migrants (net)')
sns.barplot(x='Country (or dependency)', y='Migrants (net)', data=top_5_countries)
plt.xlabel('Country')
plt.ylabel('Net Migration')
plt.title('Top 5 Countries with the Highest Net Migration')
plt.xticks(rotation=45)
plt.show()
```

```python
# Line Plot: Median Age by Country
plt.figure(figsize=(10, 6))
sns.lineplot(x='Country (or dependency)', y='Med. Age', data=data)
plt.xlabel('Country')
plt.ylabel('Median Age')
plt.title('Median Age by Country')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter Plot: Fertility Rate vs. Median Age
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Fert. Rate', y='Med. Age', data=data)
plt.xlabel('Fertility Rate')
plt.ylabel('Median Age')
plt.title('Fertility Rate vs. Median Age')
plt.show()
```

```python
# Area Plot: Cumulative Population Change over Time
plt.figure(figsize=(10, 6))
cumulative_population = data.groupby('Country (or dependency)')['Net Change'].cumsum()
data['Cumulative Population'] = cumulative_population
sns.lineplot(x='Country (or dependency)', y='Cumulative Population', data=data)
plt.xlabel('Country')
plt.ylabel('Cumulative Population Change')
plt.title('Cumulative Population Change over Time')
plt.xticks(rotation=45)
plt.show()
```

```python
# Violin Plot: Distribution of Urban Population Percentage
plt.figure(figsize=(10, 6))
sns.violinplot(data['Urban Pop %'])
plt.xlabel('Urban Population Percentage')
plt.title('Distribution of Urban Population Percentage')
plt.show()
```

```python
# Box Plot: Distribution of Net Migration
plt.figure(figsize=(10, 6))
sns.boxplot(data['Migrants (net)'])
plt.xlabel('Net Migration')
plt.title('Distribution of Net Migration')
plt.show()
```

```python
# Bar Plot: World Population Share by Country
plt.figure(figsize=(10, 6))
sns.barplot(x='Country (or dependency)', y='World Share', data=data)
plt.xlabel('Country')
plt.ylabel('World Population Share')
plt.title('World Population Share by Country')
plt.xticks(rotation=45)
plt.show()
```
